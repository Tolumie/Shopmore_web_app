The goal of this project is to segment customers of ShopMore Retail Mall using K-Means clustering. ## Step 2: Clustering Analysis Using K-Means

### Determining the Optimal Number of Clusters
To choose an appropriate number of clusters (k), we employ the Elbow Method:
- **Elbow Method:**  
  - We plot the within-cluster sum of squares (WCSS) for various cluster numbers (k values).
  - The plot shows a noticeable "elbow" at k=5, suggesting that five clusters adequately capture the structure of the data without overfitting.

### Implementing K-Means Clustering
Using k=5, we apply the K-Means algorithm to the dataset:
- **Cluster Assignment:**  
  - The algorithm assigns each customer to one of the five clusters, based on the similarity in annual income and spending score.
- **Centroids Calculation:**  
  - Centroids, representing the center of each cluster, are calculated to help interpret the cluster characteristics.

### Step 1: Import Necessary Libraries

```python
# Importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Setting plot styles for better visualizations
sns.set(style="whitegrid")
```

Explanation:
We import essential libraries: numpy and pandas for data manipulation, matplotlib and seaborn for plotting, and KMeans from sklearn for clustering. PCA is imported to reduce dimensionality for visualization.

### Step 2: Load the Dataset

```python
# Load the dataset
file_path = 'Shopmore_Retail.csv'  # Update this path to your file location
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

### Explanation:
The dataset is loaded from a CSV file named Mall_Customers.csv. We display the first few rows to get an overview of the data structure.

### Step 3: Data Exploration and Cleaning


```python
# Checking for missing values
print("Missing values in the dataset:")
print(data.isnull().sum())

# Checking for duplicates
print("\nNumber of duplicate rows in the dataset:", data.duplicated().sum())

# Summary statistics
data.describe()
```

### Insight:
There are no missing values or duplicate records in the dataset, indicating that the data is clean and ready for analysis. The summary statistics give an overview of the ranges and central tendencies of each column.


### Step 4: Exploratory Data Analysis (EDA)
### a) Visualize Data Distribution

## You can zoom in to view the chart more

```python
# Distribution of Age, Annual Income, and Spending Score
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

sns.histplot(data['Age'], bins=20, kde=True, ax=ax[0])
ax[0].set_title('Age Distribution')

sns.histplot(data['Annual Income (k$)'], bins=20, kde=True, ax=ax[1])
ax[1].set_title('Annual Income Distribution')

sns.histplot(data['Spending Score (1-100)'], bins=20, kde=True, ax=ax[2])
ax[2].set_title('Spending Score Distribution')

plt.tight_layout()
plt.show()
```

### Insight:
The histograms show that age, income, and spending scores are not evenly distributed, suggesting varied customer behaviors.

### b) Correlation Analysis

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

### Insight:
The correlation heatmap shows a weak relationship between income and spending score, indicating that high income doesn't necessarily mean high spending, motivating the need for clustering.


### Step 5: Data Preparation for Clustering


```python
# Selecting features for clustering
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]

# Displaying the first few rows of the selected features
X.head()
```

### Explanation:
We choose 'Annual Income' and 'Spending Score' for clustering since they represent financial characteristics relevant to customer segmentation.

### Step 6: Determining Optimal Number of Clusters Using the Elbow Method

```python
# Using the Elbow Method to find the optimal number of clusters
wcss = []  # List to store within-cluster sum of squares

# Calculating WCSS for different values of k
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plotting the Elbow Method results
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.show()
```

### Insight:
The plot shows a noticeable bend (elbow) at k=5, suggesting that five clusters should be chosen for the K-Means algorithm.

### Step 7: Implementing K-Means Clustering


```python
# Applying K-Means with k=5
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=42)
data['Cluster'] = kmeans.fit_predict(X)

# Save model with pickle for reuse
import pickle
pickle.dump(kmeans, open("kmeans.pkl", 'wb'))
# Displaying the first few rows with the assigned cluster labels
data.head()
```

### Explanation:
We apply K-Means clustering with five clusters. Each customer is assigned a cluster label, which is added as a new column in the dataset.

### Step 8: Visualizing the Clusters
a) 2D Scatter Plot

```python
# Visualizing the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', data=data, palette='Set1', s=100, alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='black', marker='X', label='Centroids')
plt.title('Customer Segments')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

### Insight:
The scatter plot clearly separates the five clusters, with different customer segments visible. The black 'X' markers indicate the centroids of each cluster.

### b) PCA for Dimensionality Reduction (Optional)

```python
# Applying PCA for dimensionality reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualizing clusters in the reduced dimension
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=data['Cluster'], palette='Set1', s=100, alpha=0.7)
plt.title('PCA-Reduced Cluster Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

### Explanation:
PCA helps visualize high-dimensional data in a 2D space, confirming the separation of clusters.

### 9d. Predicting new customer

```python
# Step 10: Predictive Function for New Customers
def clustering(age, avg_income, spending_score):
    kmeans_model = pickle.load(open("kmeans.pkl", 'rb'))
    new_customer = np.array([[avg_income, spending_score]])  # Updated to use income and spending score
    predicted_cluster = kmeans_model.predict(new_customer)
    cluster_mapping = {
        0: "Moderate Spenders (Age: Middle-aged, Income: Moderate, Spending: Average)",
        1: "High Spenders (Age: Young, Income: High, Spending: High)",
        2: "Cautious Savers (Age: Middle-aged, Income: High, Spending: Low)",
        3: "Young Lifestyle Spenders (Age: Youngest, Income: Low, Spending: High)",
        4: "Price-Sensitive (Age: Older, Income: Low, Spending: Low)"
    }

    return cluster_mapping.get(predicted_cluster[0], "Unknown Cluster")
```

```python
# Test Case
age = 40
avg_income = 30  # Changed from avg_spend to avg_income
spending_score = 4  # Assuming visit_per_week and promotion_interest are not used in prediction

predicted_cluster = clustering(age, avg_income, spending_score)
print(f'New Customer assigned to Cluster: {predicted_cluster}')
```
